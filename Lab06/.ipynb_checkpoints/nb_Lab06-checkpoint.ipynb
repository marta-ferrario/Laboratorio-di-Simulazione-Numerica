## Esercizio 6.1

In questo esercizio viene studiato il modello di Ising monodimensionale: gli spin di un sistema non sono indipendenti ma interagiscono con i vicini più prossimi; assumiamo che l'interazione sia $J=1$, quindi il sistema favorisce l'allineamento degli spin, e che siano presenti $N=50$ spin. Il sistema è descritto dall’Hamiltoniana del modello di Ising e viene campionato secondo la distribuzione di Boltzmann $p({s_i})= \frac{e^{-\beta H}}{Z}$.

Per generare le configurazioni di spin si utilizzano due algoritmi Monte Carlo: il metodo di Metropolis e il campionamento di Gibbs. L’obiettivo è confrontare i due metodi e studiare l’andamento delle principali grandezze termodinamiche al variare della temperatura nell’intervallo $T \in [0.5, 2.0]$. 

Le osservabili calcolate sono: 
- energia interna $U = \braket{H}$
- capacità termica $ C = \beta^2 (\braket{H^2} - \braket{H}^2)$
- suscettività magnetica $\chi = \beta \braket{(\sum_i s_i)^2}$
- magnetizzazione $M = \braket{\sum_{i=1}s_i}$ in presenza di campo magnetico esterno $h=0.02$.

Tutte le quantità sono stimate utilizzando il metodo del data blocking per ottenere una corretta valutazione delle incertezze statistiche.

```python
import numpy as np
import matplotlib.pyplot as plt

# LETTURA DATI DA FILE
chi_gibbs = np.loadtxt("NSL_SIMULATOR/OUTPUT/PLOT/chi_gibbs.dat")
chi_metro = np.loadtxt("NSL_SIMULATOR/OUTPUT/PLOT/chi_metro.dat")


t_gibbs = chi_gibbs[:,0]
t_metro = chi_metro[:,0]
av_gibbs = chi_gibbs[:,2]
av_metro = chi_metro[:,2]
err_gibbs = chi_gibbs[:,3]
err_metro = chi_metro[:,3]


# TEORICO
J = 1.0
Ns = 50
th = np.tanh(J/t_gibbs)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
beta = 1/(t_gibbs)
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

# PLOT
fig, axes = plt.subplots(1, 2, figsize=(12,5))

# --- Gibbs ---
axes[0].errorbar(t_gibbs, av_gibbs, yerr=err_gibbs,
                 fmt='o', markersize=4,
                 color='blue', label="Gibbs", capsize=2)

axes[0].plot(t_gibbs, X, color='lightgreen', label="Teorico")

axes[0].set_xlabel("T")
axes[0].set_ylabel("Chi")
axes[0].set_title("Suscettività (Gibbs)")
axes[0].legend()
axes[0].grid(True)

# --- Metropolis ---
axes[1].errorbar(t_metro, av_metro, yerr=err_metro,
                 fmt='o', markersize=4,
                 color='orange', label="Metropolis", capsize=2)

axes[1].plot(t_metro, X, color='lightgreen', label="Teorico")

axes[1].set_xlabel("T")
axes[1].set_ylabel("Chi")
axes[1].set_title("Suscettività (Metropolis)")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# LETTURA DATI DA FILE
cv_gibbs = np.loadtxt("NSL_SIMULATOR/OUTPUT/cv_gibbs.dat")
cv_metro = np.loadtxt("NSL_SIMULATOR/OUTPUT/cv_metro.dat")


t_gibbs = cv_gibbs[:,0]
t_metro = cv_metro[:,0]
av_gibbs = cv_gibbs[:,2]
av_metro = cv_metro[:,2]
err_gibbs = cv_gibbs[:,3]
err_metro = cv_metro[:,3]


# TEORICO
J = 1.0
Ns = 50
th = np.tanh(J/t_gibbs)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
beta = 1/(t_gibbs)
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

# PLOT
plt.figure(figsize=(8,5))
plt.errorbar(t_gibbs, av_gibbs, yerr=err_gibbs, fmt='o', markersize=3, label="Gibbs", capsize=2)
plt.errorbar(t_metro, av_metro, yerr=err_metro, fmt = 'o', markersize=3, label="Metro", capsize=2)
plt.plot(t_gibbs, heat, 'r-', label="Teorico")

plt.xlabel("T")
plt.ylabel("Cv")
plt.title("Capacità termica")
plt.legend()
plt.grid(True)


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# LETTURA DATI DA FILE
magnet_gibbs = np.loadtxt("NSL_SIMULATOR/OUTPUT/PLOT/magnet_gibbs.dat")
magnet_metro = np.loadtxt("NSL_SIMULATOR/OUTPUT/PLOT/magnet_metro.dat")


t_gibbs = magnet_gibbs[:,0]
t_metro = magnet_metro[:,0]
av_gibbs = magnet_gibbs[:,2]
av_metro = magnet_metro[:,2]
err_gibbs = magnet_gibbs[:,3]
err_metro = magnet_metro[:,3]


# TEORICO
J = 1.0
Ns = 50
th = np.tanh(J/t_gibbs)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
b = 1/(t_gibbs)
h=0.02 #external field

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

# PLOT
fig, axes = plt.subplots(1, 2, figsize=(12,5))

# --- Gibbs ---
axes[0].errorbar(t_gibbs, av_gibbs, yerr=err_gibbs,
                 fmt='o', markersize=3,
                 color='blue', label="Gibbs", capsize=2)

axes[0].plot(t_gibbs, M, color='lightgreen', label="Teorico")

axes[0].set_xlabel("T")
axes[0].set_ylabel("Magnetizzazione")
axes[0].set_title("Magnetizzazione (Gibbs)")
axes[0].legend()
axes[0].grid(True)

# --- Metropolis ---
axes[1].errorbar(t_metro, av_metro, yerr=err_metro,
                 fmt='o', markersize=3,
                 color='orange', label="Metropolis", capsize=2)

axes[1].plot(t_gibbs, M, color='lightgreen', label="Teorico")

axes[1].set_xlabel("T")
axes[1].set_ylabel("Magnetizzazione")
axes[1].set_title("Magnetizzazione (Metropolis)")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
