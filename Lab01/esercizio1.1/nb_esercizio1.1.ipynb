```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("data.dat")

x=data[:,0]
y=data[:,1]-0.5
error=data[:,2]

plt.plot(x,y, ".")
plt.xlabel("Lanci")
plt.ylabel("<r> - 0.5")
plt.errorbar(x,y,yerr=error,)
plt.grid(True)
plt.title("Valore medio")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

chi2 = np.loadtxt("data_chi.dat")
plt.hist(chi2, bins=20)

plt.xlabel("chi quadro")
plt.ylabel("frequenza")
plt.title("Distribuzione dei valori di chi quadro")

plt.show()

print("media =", np.mean(chi2))
print("deviazione standard =", np.std(chi2))
```
