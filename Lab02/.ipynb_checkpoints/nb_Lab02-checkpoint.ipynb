## Esercizio 2.1
#### INTRODUZIONE
In questo esercizio si vuole calcolare numericamento il seguente integrale: 

$$ I = \int_0^1 \frac{\pi}{2} \cos{\left ( \frac{\pi x}{2} \right)} \, dx =1 $$

Per fare ciò, vengono utilizzati due metodi differenti: 
1. **Sampling uniform in $[0,1]$**: l'integrale è quindi dato da $I \sim \frac{1}{M} \sum_{i=1}^M f(x_i)$, dove $x_i$ sono i punti estratti e $M$ è il numero totale di estrazioni. 
2. **Importance sampling**: si sceglie una distribuzione di probabilità non uniforme $p(x)$ più adatta alla funzione integranda. Si riscrive dunque l'integrale come: $$I \sim \frac{1}{M} \sum_{i=1}^M \frac{f(x_i)}{p(x_i)}$$ Per determinare la distribuzione di probabilità $p(x)$, si prende inizialmente la generica retta di equazione $y=ax+b$ e si impone che si annulli nello stesso punto della funzione $f(x)$ e la normalizzazione. Svolgendo pochi passaggi algebrici, si ricava che: $$p(x)= 2(1-x)$$ Per generare variabili causali distribuite secondo $p(x)= 2(1-x)$ è stato utilizzato il metodo della trasformazione inversa. Calcolando la funzione cumulativa $F(x)= \int_0^x p(t) \, dt = 2x -x^2$ e imponendo $u=F(x)$, con $u$ estratto uniformemente in $[0,1]$, si ottiene: $$ x = 1- \sqrt{1-u}$$ che permette di trasformare numeri uniformente distribuiti secondo la distribuzione di probabilità desiderata. Di seguito sono riportati i grafici dell'integrando e della distribuzione di probabilità scelti. 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 1000)

# integrando
f = (np.pi/2) * np.cos(np.pi * x / 2)
# pdf importance sampling
p = 2 * (1 - x)

# plot
plt.figure(figsize=(8,5))
plt.plot(x, f, label=r'$f(x)=\frac{\pi}{2}\cos\left(\frac{\pi x}{2}\right)$', color = 'navy')
plt.plot(x, p, '--', label=r'$p(x)=2(1-x)$', color = 'darkorange')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Confronto integrando e $p(x)$')

plt.grid(True)
plt.legend()

plt.show()
```

#### STRUTTURA CODICE C++ E ANALISI DATI

L'esercizio è stato implementato in C++ utilizzando il metodo a blocchi, con numero totale di lanci $M=10^4$ e numero di blocchi $N=100$. Nel grafico riportato qui sotto è presente un confronto del valore dell'integrale meno il valore teorico ($I=0$) calcolato con il metodo del sampling uniforme e dell'importance sampling. 

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("esercizio2.1/data.dat")
data_imp = np.loadtxt("esercizio2.1/data_imp.dat")

x=data[:,0]
y=data[:,1]-1
y_imp=data_imp[:,1]-1
error=data[:,2]
error_imp=data_imp[:,2]

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# PLOT VALOR MEDIO
axes[0].plot(x,y, ".", label = 'Sampling uniforme')
axes[0].plot(x,y_imp, ".", label = 'Importance sampling')
axes[0].errorbar(x,y,yerr=error, color="navy", linestyle="none")
axes[0].errorbar(x,y_imp,yerr=error_imp, color="darkorange", linestyle="none")
axes[0].set_xlabel("Blocchi")
axes[0].set_ylabel("$I-I_{teorico}$")
axes[0].set_title("$I-I_{teorico}$")
axes[0].grid(True)
axes[0].legend(); 

# PLOT ERRORE
axes[1].errorbar(x, error, marker='', linestyle='-', color='navy', label = 'Sampling uniforme')
axes[1].errorbar(x, error_imp, marker='', linestyle='-', color='darkorange' , label = 'Importance sampling')
axes[1].set_xlabel("Blocchi")
axes[1].set_ylabel("Errore")
axes[1].set_title("Errore")
axes[1].grid(True)
axes[1].legend(); 

plt.tight_layout()
plt.show()

plt.show()
```

Dai grafici, si osserva che entrambi i metodi convergono al valore teorico. Tuttavia, nel caso di sampling uniforme la convergenza è più lenta, le fluttuazioni sono più grandi e l'errore statistico è maggiore. Viceversa, il grafico dell'importance sampling mostra una convergenza più rapida e incertezze significamente più piccole. Nel grafico dell'errore si osserva che in entrambi i casi questo scala come $\sim \frac{1}{\sqrt{N}}$

In conclusione, si è visto come la scelta di una distribuzione $p(x)$ simile alla forma della funzione integranda migliora l’efficienza del metodo.

## Esercizio 2.2
#### INTRODUZIONE
In questo esercizio viene studiato il comportamento di una Random Walk tridimensionale, ovvero un processo stocastico in cui una particella si muove effettuando una successione di passi casuali nello spazio. L’obiettivo è simulare numericamente molte traiettorie indipendenti che partono tutte dall’origine e analizzare come cresce la distanza media percorsa al variare del numero di passi.

Sono stati considerati due casi distinti: 
1. **Random Walk su reticolo cubico (cubic lattice)**: ad ogni passo la particella si muove di una distanza unitaria a=1 lungo una delle tre direzioni principali dello spazio (x, y, z), scegliendo casualmente sia l’asse sia il verso del movimento.
2. **Random Walk nel continuo**: ad ogni passo la particella si sposta sempre di una distanza unitaria, ma lungo una direzione casuale nello spazio, ottenuta campionando uniformemente l’angolo solido.

Per entrambe le simulazioni è stata calcolata la quantità: $$\sqrt{\braket{r_N^2}}$$ dove $r_N^2= x_N^2 + y_N^2 + z_N^2$ rappresenta il quadrato della distanza dall’origine dopo N passi e $\braket{\cdot}$ indica la media effettuata su molte realizzazioni della random walk.

#### STRUTTURA CODICE C++
Per eseguire la simulazione è stato sviluppato un codice C++ che sfrutta il generatore di numeri casuali fornito. La simulazione è stata organizzata secondo il metodo a blocchi, utile per stimare correttamente le incertezze statistiche. Sono stati utilizzati i seguenti parametri: 
- Numero totale di random walk: $M=10000$
- Numero di passi per ogni traiettoria: $N=100$
- Numero di blocchi: $N_{blocchi}=100$

Una volta calcolata la grandezza richiesta $\sqrt{\braket{r_N^2}}$, è necessario calcolare l'incertezza statisica. Quest'ultima è stata ottenuta tramite propagazione dell'errore: 

$$ \sigma_{\sqrt{x}} = \frac{1}{2\sqrt{x}} \sigma_x \qquad \text{con } \sigma_x = \sqrt{\frac{\braket{x^2} - \braket{x^2}}{N_{blocchi}}}$$


#### ANALISI DEI RISULTATI
Per entrambi i tipi di random walk sono stati costruiti due grafici:
- Andamento di $\sqrt{\braket{r_N^2}}$ in funzione del numero di passi riportando anche le barre d’errore statistiche; quest'ultime sono state amplificate di un fattore 10 per renderle visibili.
- Andamento dell'errore in funzione del numero di passi. 

In entrambi i casi si osserva una crescita della distanza media proporzionale a $\sqrt{N}$, come previsto teoricamente per un processo diffusivo. Il comportamento nella random walk nel continuo è molto simile a quello osservato nel reticolo cubico: nonostante il diverso modo di generare il moto, il processo mantiene le caratteristiche tipiche della diffusione.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("esercizio2.2/data_lattice.dat")

x = data[:, 0]
y = data[:, 1]
error = data[:, 2]

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# PLOT VALOR MEDIO
axes[0].plot(x, y, ".", color="navy", label = 'RW reticolo cubico')
axes[0].errorbar(x, y, yerr=error*10, color="navy", linestyle="none")
axes[0].set_xlabel("Passi")
axes[0].set_ylabel(r"$\sqrt{\langle \vec{r}_N\ ^2 \rangle}$")
axes[0].set_title("Distanza media percorsa")
axes[0].grid(True)
axes[0].legend()

# PLOT ERRORE
axes[1].errorbar(x, error, marker='', linestyle='-', color='darkorange')
axes[1].set_xlabel("Passi")
axes[1].set_ylabel("Errore")
axes[1].set_title("Errore")
axes[1].grid(True)

plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("esercizio2.2/data_cont.dat")

x = data[:, 0]
y = data[:, 1]
error = data[:, 2]

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# PLOT VALOR MEDIO
axes[0].plot(x, y, ".", color="forestgreen", label = 'RW continuo')
axes[0].errorbar(x, y, yerr=error*10, color="forestgreen", linestyle="none")
axes[0].set_xlabel("Passi")
axes[0].set_ylabel(r"$\sqrt{\langle \vec{r}_N\ ^2 \rangle}$")
axes[0].set_title("Distanza media percorsa")
axes[0].grid(True)
axes[0].legend()

# PLOT ERRORE
axes[1].errorbar(x, error, marker='', linestyle='-', color='gold')
axes[1].set_xlabel("Passi")
axes[1].set_ylabel("Errore")
axes[1].set_title("Errore")
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
